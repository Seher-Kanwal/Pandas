```python
# we Will change it into actuall date
df = pd.read_csv('weather_data.csv', parse_dates = ['day'])
df
type(df.day[0])
```

```python
# set day as an index, we have to use inplace argument
df.set_index('day',inplace = True)
df
```

```python
# it will fill all the missing values with a single value given in pranthesies
df1 = df.fillna(0)
df1
```

```python
#in order to fill out different columns with different values we create a dict
df1= df.fillna({
    'temperature': 0,
    'windspeed': 3.4,
    'event':"no event"
})
df1
```

```python
# in order to fill above values in missing cells we used forward fill 
# for filling the below values we used below fill
#df1 = df.fillna(method = 'ffill')
df1 = df.fillna(method = 'bfill')
df1
```

```python
# in order to put limit that only one missing value will change 
# we need to use limit
df1 = df.fillna(method = 'bfill' , limit = 1)
df1
```

```python
df.describe()
```

```python
# interpolate is by deafult is linear, it finds the mid between two values and fill the middle missing values with the answer
# mid of 1, 3 is 30 so 2nd is filled with 30
df1 = df.interpolate()
df1
```

```python
# but if look at the date which is more closer to the third one
# for more acurate result we will use another introplate method time
df1 = df.interpolate(method = 'time')
df1
```

```python
# it will drop all the rows that contains NAN values
df1 = df.dropna()
df1
```

```python
# it will only drop that row that contain NaN values in all the cells
df1 = df.dropna(how = 'all')
df1
```

```python
# we can define a threshhold for droppping the rows
# it will not drop the rows that have 2 valid values in the row
df1= df.dropna(thresh=2)
df1
```
